```python
import requests
import json
import FinanceDatabase as fd
import pandas as pd
from pandas import json_normalize 


headers = {'x-api-key':"d5I7gNKqdR1kaqxNy4XXl1dvsMBjLIHHaJVp7Rji"}
# symbol = 'AAPL'

# url = f'https://yfapi.net/v11/finance/quoteSummary/{symbol}'

# querystring = {'modules' : 'defaultKeyStatistics'}
# response = requests.request("GET", url, headers=headers, params=querystring)
# print(response)
# dict_response = json.loads(response.text)["quoteSummary"]['result'][0]['defaultKeyStatistics']

# # data = pd.DataFrame(dict_response, index =[0])
# data = pd.DataFrame.from_dict(dict_response)
# # print(data)
# data.to_excel('data.xlsx', index=False)




# regularMarketPrice = get_quote("AAPL", "price")["regularMarketPrice"]["raw"]
# totalCurrentAssets = get_quote("AAPL", "balanceSheetHistoryQuarterly")["balanceSheetStatements"][0]["totalCurrentAssets"]["raw"]
# totalLiab = get_quote("AAPL", "balanceSheetHistoryQuarterly")["balanceSheetStatements"][0]["totalLiab"]["raw"]
# sharesOutstanding = get_quote("AAPL", "defaultKeyStatistics")["sharesOutstanding"]["raw"]
# ncav = totalCurrentAssets - totalLiab

# ncavs = ncav / sharesOutstanding


# ### 야후 파이낸스 유료 API 불러오기 함수 ###
def get_quote(symbol, modules):
    url = f'https://yfapi.net/v11/finance/quoteSummary/{symbol}'
    querystring = {"symbols": symbol, "modules": modules}
    response = requests.request("GET", url, headers=headers, params=querystring)
    dict_response = json.loads(response.text)
    return dict_response["quoteSummary"]["result"][0][modules]


#1. symbol list 저장
stocks = fd.select_equities(country='United States')
symbol_lists = list(stocks.keys())


#2. ncav_list 생성 및 dataframe으로 형식 지정 
ncav_list =[]
#Create a DataFrame object
df = pd.DataFrame(ncav_list, columns = ['regularMarketPrice', 'totalCurrentAssets', 'totalLiab', 'sharesOutstanding', 'ncav', 'ncavs'])

#3. yahooapi에서 data 불러와서 ncav 관련 data 저장

for i in symbol_lists:
    try:
        regularMarketPrice = get_quote(i, "price")["regularMarketPrice"]["raw"]
        totalCurrentAssets = get_quote(i, "balanceSheetHistoryQuarterly")["balanceSheetStatements"][0]["totalCurrentAssets"]["raw"]
        totalLiab = get_quote(i, "balanceSheetHistoryQuarterly")["balanceSheetStatements"][0]["totalLiab"]["raw"]
        sharesOutstanding = get_quote(i, "defaultKeyStatistics")["sharesOutstanding"]["raw"]
        ncav = totalCurrentAssets - totalLiab
        ncavs = ncav / sharesOutstanding
        df.loc[i] = [regularMarketPrice, totalCurrentAssets, totalLiab, sharesOutstanding, ncav, ncavs]
    except:
        pass

df.to_excel("data.xlsx")

#### 각각함수따로 했을때 ####
# def marketCap(symbol, modules):
    
#     querystring = {"symbols": symbol, "modules": modules}
#     response = requests.request("GET", url, headers=headers, params=querystring)
#     dict_response = json.loads(response.text)
#     marketCap = dict_response["quoteSummary"]["result"][0]["summaryDetail"]["marketCap"]["raw"]
#     return marketCap

# def regularMarketPrice(symbol, modules):
#     querystring = {"symbols": symbol, "modules": modules}
#     response = requests.request("GET", url, headers=headers, params=querystring)
#     dict_response = json.loads(response.text)
#     regularMarketPrice = dict_response["quoteSummary"]["result"][0]["price"]["regularMarketPrice"]["raw"]
#     print(regularMarketPrice)
# #marketcap 구하기

# regularMarketPrice()

# def ncav(symbol, modules):
#     querystring = {"symbols": symbol, "modules": modules}
#     response = requests.request("GET", url, headers=headers, params=querystring)
#     dict_response = json.loads(response.text)
#     totalCurrentAssets = dict_response["quoteSummary"]["result"][0]["balanceSheetHistoryQuarterly"]["balanceSheetStatements"][0]["totalCurrentAssets"]["raw"]
#     totalLiab = dict_response["quoteSummary"]["result"][0]["balanceSheetHistoryQuarterly"]["balanceSheetStatements"][0]["totalLiab"]["raw"]
#     ncav = totalCurrentAssets - totalLiab
#     return ncav

# # print(ncav - marketCap )

# def sharesOutstanding(symbol, modules):
#     querystring = {"symbols": symbol, "modules": modules}
#     response = requests.request("GET", url, headers=headers, params=querystring)
#     dict_response = json.loads(response.text)
#     sharesOutstanding = dict_response["quoteSummary"]["result"][0]["defaultKeyStatistics"]["sharesOutstanding"]["raw"]
#     return sharesOutstanding


# marketCap("AAPL", "summaryDetail")

```
